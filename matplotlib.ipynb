```python
# Linear function
# y = 2x + 1 or  y = x^2

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 1, 50)
# y = 2*x + 1
y = x**2
plt.plot(x, y)
plt.show()
```

```python
# Separate to two figure (Windowlize in py form)
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3, 3, 50)
y1 = 2*x + 1
y2 = x**2

plt.figure()
plt.title('figure 1')
plt.plot(x, y1)

plt.figure(num=3, figsize=(8, 5))
plt.title('figure 2')
plt.plot(x, y2)
plt.plot(x, y1, color='red', linewidth=1.0, linestyle='--')
plt.show()
```

```python
# Setting name by Axis label(title)
x = np.linspace(-3, 3, 50)
y1 = 2*x + 1
y2 = x**2

plt.figure()
plt.plot(x, y2)
plt.plot(x, y1, color='red', linewidth=1.0, linestyle='--')


# X & Y axis numeric range (取值范围) & Label name 命名
plt.xlim((-1, 2))
plt.ylim((-2, 3))
plt.xlabel('I am x')
plt.ylabel('I am y')

# substitute a new X or Y label
new_ticks = np.linspace(-1, 2, 5)
print(new_ticks)
plt.xticks(new_ticks)
plt.yticks([-2, -1.8, -1, 1.22, 3,],
           [r'$really\ bad$', r'$bad\ \alpha$', r'$normal$', r'$good$', r'$really\ good$'])
          #  ['really bad', 'bad', 'normal', 'good', 'really good']
plt.show()
```

```python
# Setting name by Axis label(title)
x = np.linspace(-3, 3, 50)
y1 = 2*x + 1
y2 = x**2

plt.figure()
plt.plot(x, y2)
plt.plot(x, y1, color='red', linewidth=1.0, linestyle='--')


# X & Y axis numeric range (取值范围) & Label name 命名
plt.xlim((-1, 2))
plt.ylim((-2, 3))
plt.xlabel('I am x')
plt.ylabel('I am y')

# substitute a new X or Y label
new_ticks = np.linspace(-1, 2, 5)
print(new_ticks)
plt.xticks(new_ticks)
plt.yticks([-2, -1.8, -1, 1.22, 3,],
           [r'$really\ bad$', r'$bad\ \alpha$', r'$normal$', r'$good$', r'$really\ good$'])

# gca  = 'get current axes' (子图)
# spines(图-边框线)
ax = plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.xaxis.set_ticks_position('bottom') # 指定设置x轴(y轴) >> bottom代替x axis
ax.yaxis.set_ticks_position('left')
ax.spines['bottom'].set_position(('data', 0)) # 挪动x&y轴位置; x轴原点放在指定的y值上
ax.spines['left'].set_position(('data', 0)) # parameters: 1.outward,  2.axes
plt.show()
```

```python
# Legend 图例
x = np.linspace(-3, 3, 50)
y1 = 2*x + 1
y2 = x**2

plt.figure()

plt.xlim((-1, 2))
plt.ylim((-2, 3))
plt.xlabel('I am x')
plt.ylabel('I am y')


new_ticks = np.linspace(-1, 2, 5)
print(new_ticks)
plt.xticks(new_ticks)
plt.yticks([-2, -1.8, -1, 1.22, 3,],
           [r'$really\ bad$', r'$bad\ \alpha$', r'$normal$', r'$good$', r'$really\ good$'])

L1, = plt.plot(x, y2, label='up')
L2, = plt.plot(x, y1, color='red', linewidth=1.0, linestyle='--', label='down')
# plt.legend(loc='best') # call
plt.legend(handles=[L1,L2,], labels=['aaa', 'bbb'], loc='best')
plt.show()
```
